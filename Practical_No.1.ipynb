**5. Python Program for compound interest**

```python
p = float(input("Enter principal:  "))
t = float(input("Enter No. of days:  "))
r = float(input("Enter rate on interest:  "))

Compound_interest =  p * (pow((1 + r / 100), t))          #Formula for componund  intrest
print('The compound interest on provided values is: ', Compound_interest)
```

**6. Python Program to check Armstrong number**

```python

num = int(input("Enter a number: "))        # take input from the user
sum = 0
temp = num

while temp > 0:
   r = temp % 10
   sum = sum + r ** 3
   temp //= 10                             # Formula to find out armstrong number

if num == sum:                             # display the result
   print("Yess, it's an Armstrong number")
else:
   print("No, it's not an Armstrong number")
```

**8. Python program to print all prime numbers in an interval**

```python
lower = 1
upper = 10

print('The prime numbers between 1 to 10 are : ')
for num in range(lower , upper+1):
  if num > 1:
      for i in range(2, num):
          if (num % i) == 0:
             break
      else:
            print(num)    
```

**9. Python program to check whether a number is Prime or not**

```python
num = int(input("Enter the value to cheeck whether the number prime of not: "))


for i in range(2, num-1):
  if(num % i) == 0:
      print("The entered number is not prime number")
      break
  else:
      print("The entered number is prime number")  
      break
```

**15. Python program for cube sum of first n natural numbers**

```python

def cube(n):
    s=0
    for i in range(n+1):
        s+=i**3
    return s

n=int(input("enter n: "))                       #input
print("sum of cubes of first  natural numbers: ".format(n),cube(n))
```
